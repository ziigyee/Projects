### Page 1 to 20 , click Next, add to counter. if counter = 20, then driver.close()

```python
search_terms = ['data']  #['data analyst','data scientist', 'business analyst']
```

```python
url = 'https://www.jobstreet.com.sg/en/job-search/job-vacancy.php?key={}'.format(search_terms[0])
job_links = []
print('Page {}'.format(url))  
page_counter = 1
```

```python
def get_source_code(page_source):
    html = BeautifulSoup(page_source, "lxml")
    job_url =  html.find_all('a', attrs={'class':'position-title-link'}, href=True)

    if len(job_url) < 1:
        print('no job URLS..')
    else:
        print('ok. got {} job urls'.format(len(job_url)))
        # add to our quotes list
        for jd in job_url:
            with open("urls.txt", "a") as f:
                f.write(jd['href']+',')
                job_links.append(jd['href']+',')
```
